2. 쯈u칠 es un RDD en Spark? Describe una de sus principales caracter칤sticas. 쯈u칠 tienen que ver con los dataframes?.

Un RDD (Resilient Distributed Dataset) es la abstracci칩n de datos primaria en Spark y proporciona una forma inmutable y distribuida de manejar los datos, permitiendo el procesamiento en m칰ltiples nodos de un cl칰ster. Los RDD son resistentes, lo que significa que son capaces de recomputar datos en caso de fallas.

Caracter칤sticas de los RDD:

La resiliencia es una de sus caracter칤sticas definitorias. Esto se logra a trav칠s de la propiedad de tolerancia a fallos, que permite que partes del dataset que se hayan perdido en el caso de un fallo en el nodo se reconstruyan autom치ticamente. Los RDD logran esta resiliencia almacenando el linaje de las transformaciones aplicadas a los datos iniciales, permitiendo una reconstrucci칩n eficiente de los mismos.

Inmutabilidad. Una vez que se crea un RDD, no se puede modificar. Cualquier operaci칩n que modifique un RDD, como map o filter, resultar치 en un nuevo RDD.

Los datos en un RDD est치n distribuidos y paralelizados autom치ticamente a trav칠s de los nodos del cl칰ster.

Lazy Evaluation. Spark no ejecuta las transformaciones de un RDD hasta que se realiza una acci칩n que requiera que un resultado sea devuelto al driver.

Los RDD pueden ser almacenados en memoria durante el procesamiento, lo que permite un acceso m치s r치pido, especialmente 칰til para aplicaciones que acceden varias veces a los mismos datos.

Relaci칩n entre RDD y DataFrames:

Los DataFrames en Spark son una abstracci칩n m치s alta que los RDD, dise침ada para simplificar el trabajo con grandes conjuntos de datos estructurados y semiestructurados mediante una interfaz similar a la de los DataFrames de Pandas pero distribuida. Un DataFrame en Spark representa una tabla de datos con filas y columnas, donde cada columna tiene un nombre y un tipo de dato.

Los DataFrames est치n construidos sobre el concepto de RDDs, utilizando un RDD de filas (Row RDD) donde cada fila es un objeto que representa una fila l칩gica del DataFrame. Los DataFrames proporcionan optimizaciones a trav칠s del Catalyst query optimizer, que construye un plan de ejecuci칩n eficiente para las consultas, algo que los RDD no hacen autom치ticamente. Adem치s, los DataFrames ofrecen una API muy util para la manipulaci칩n de datos, permitiendo integraci칩n con sistemas SQL, lo cual facilita mucho el trabajo de transformaci칩n y an치lisis de datos estructurados.

3. Diferencia entre transformaciones y acciones en Spark. Proporciona ejemplos de cada una. 쯈u칠 ocurre internamente cuando se ejecuta una acci칩n?

Las transformaciones son operaciones en Spark que crean un nuevo RDD a partir de uno existente. Son lazy, es decir, no se computan inmediatamente cuando se definen, Spark construye un DAG (Directed Acyclic Graph), que es un plan de ejecuci칩n que describe todas las transformaciones que deben ser aplicadas al RDD. Las transformaciones solo se ejecutan cuando una acci칩n se lleva a cabo.

Ejemplos:

```python
map(func): Retorna un nuevo RDD aplicando una funci칩n a cada elemento del RDD original.

rdd = sc.parallelize([1, 2, 3, 4])
rdd2 = rdd.map(lambda x: x * 2)

filter(func): Retorna un nuevo RDD conteniendo solo los elementos que satisfacen una condici칩n.

rdd = sc.parallelize([1, 2, 3, 4])
filtered_rdd = rdd.filter(lambda x: x > 2)

flatMap(func): Similar a map, pero cada elemento de entrada puede ser mapeado a 0 o m치s elementos de salida (es decir, "aplanar" la lista).

rdd = sc.parallelize(["hello world", "how are you"])
rdd2 = rdd.flatMap(lambda x: x.split(" "))
```

Las acciones son operaciones que devuelven un resultado final del procesamiento del dataset. A diferencia de las transformaciones, cuando se ejecuta una acci칩n, Spark ejecuta el DAG creado por las transformaciones para entregar un resultado. Las acciones son necesarias para materializar los resultados del procesamiento y pueden desencadenar la computaci칩n del RDD.

Ejemplos de Acciones:

```python
collect(): Devuelve todos los elementos del RDD como un arreglo al programa driver.

rdd = sc.parallelize([1, 2, 3, 4])
result = rdd.collect()

count(): Retorna el n칰mero de elementos en el RDD.

rdd = sc.parallelize([1, 2, 3, 4])
count = rdd.count()

take(n): Devuelve los primeros n elementos del RDD.

rdd = sc.parallelize([1, 2, 3, 4])
first_three = rdd.take(3)

reduce(func): Agrega los elementos del RDD juntos utilizando una funci칩n asociativa y conmutativa.

rdd = sc.parallelize([1, 2, 3, 4])
sum_result = rdd.reduce(lambda x, y: x + y)
```

Cuando se ejecuta una acci칩n, Spark computa el RDD y sus dependencias que han sido definidos por las transformaciones:

- Eval칰a el DAG para entender las dependencias entre las diferentes transformaciones y RDDs.
- Optimiza el plan de ejecuci칩n, puede combinar operaciones.
- Distribuye la ejecuci칩n del DAG a trav칠s de los nodos en el cl칰ster para realizar el procesamiento en paralelo.
- Gestiona la recuperaci칩n de fallos al recomputar partes perdidas del DAG si es necesario.
- Retorna los resultados al driver o escribe los resultados a un sistema de almacenamiento externo, dependiendo de la acci칩n especificada.

El proceso de Laza evaluation y ejecuci칩n basada en acciones permite que Spark maneje eficientemente grandes vol칰menes de datos con un uso 칩ptimo de recursos. Esto tambi칠n ayuda a evitar repetidos c칩mputos innecesarios y permite a Spark ejecutar optimizaciones importantes en el plan de procesamiento.

4. Explica la importancia del particionamiento en Spark y c칩mo afecta el rendimiento del procesamiento de datos.

El particionamiento es util para el rendimiento en el procesamiento distribuido de datos. Afecta directamente c칩mo los datos son distribuidos a trav칠s de los nodos del cl칰ster y tiene un impacto significativo en la eficiencia de las operaciones de procesamiento paralelo.

Determina el nivel de paralelismo en las operaciones de Spark. Cada partici칩n de un RDD, DataFrame o Dataset puede ser procesada en paralelo por diferentes nodos del cl칰ster. Si los datos est치n bien particionados, esto puede llevar a una utilizaci칩n m치s eficiente de los recursos del cl칰ster y a una ejecuci칩n m치s r치pida de las operaciones.

Un particionamiento eficaz asegura que el trabajo sea equitativamente distribuido entre todos los nodos del cl칰ster. Esto evita que algunos nodos est칠n sobrecargados mientras que otros est칠n subutilizados, lo que puede llevar a cuellos de botella y reducir el rendimiento general del sistema.

Durante ciertas operaciones, como los joins, groupBy o reduceByKey, los datos pueden necesitar ser redistribuidos (shuffled) entre diferentes nodos para agrupar los elementos necesarios para la operaci칩n. El shuffle es una operaci칩n costosa porque involucra la transferencia de datos a trav칠s de la red y puede ser un factor limitante de rendimiento. Un buen particionamiento puede reducir la cantidad de datos que necesitan ser movidos, minimizando as칤 el costo asociado con el shuffle.

EL particionamiento puede aumentar la localidad de los datos, lo que significa que los datos necesarios para procesar una tarea est치n f칤sicamente m치s cerca del proceso que los necesita (idealmente en la misma m치quina). Esto reduce la latencia y el costo de acceso a los datos, mejorando el rendimiento de las operaciones de lectura y escritura.

Para operaciones que no requieren un shuffle, como map() o filter(), un particionamiento bien dise침ado asegura que el procesamiento se distribuya de manera uniforme, lo que maximiza el uso del paralelismo y reduce el tiempo total de ejecuci칩n.

Para operaciones que involucran un shuffle, un particionamiento previo que alinee los datos de acuerdo con la clave sobre la cual se realizar치 la operaci칩n (por ejemplo, usando partitionBy en un PairRDD) puede reducir significativamente la cantidad de datos que deben ser transferidos entre nodos.

Se controlar el particionamiento a trav칠s de varias configuraciones y m칠todos como repartition(), coalesce() o partitionBy()

5. 쮺u치les son las funciones de Spark Driver y Spark Executor?abs

El Driver es el proceso responsable de la creaci칩n del contexto de Spark, la conversi칩n del programa del usuario en tareas, y la planificaci칩n de estas tareas en los executors. El Driver convierte la aplicaci칩n Spark, que puede ser escrita en Scala, Python o Java, en un plan de ejecuci칩n f칤sico que se puede ejecutar a trav칠s del cl칰ster. Esto implica la construcci칩n del DAG (Directed Acyclic Graph) de las tareas a realizar. Programa las tareas individuales que deben ejecutarse, bas치ndose en el DAG. Luego, env칤a estas tareas a los executors apropiados para su ejecuci칩n. Maneja las sesiones de Spark, incluyendo la configuraci칩n y mantenimiento del SparkContext (en Spark 1.x) o SparkSession (en Spark 2.x y posteriores). En algunos modos de despliegue, como en standalone o Mesos, el Driver tambi칠n puede ser responsable de la gesti칩n de recursos, negociando con el gestor de cl칰ster para adquirir recursos para los executors. Una vez que los executors completan las tareas, env칤an los resultados de vuelta al Driver. Este puede entonces compilar los resultados de todas las tareas para formar la salida final de la aplicaci칩n, o realizar m치s transformaciones y acciones seg칰n sea necesario.

Los Executors son procesos lanzados para la duraci칩n de la aplicaci칩n Spark en los nodos del cl칰ster, que ejecutan las tareas asignadas y devuelven los resultados al Driver. Sson responsables de ejecutar las tareas que les asigna el Driver. Cada executor puede ejecutar m칰ltiples tareas en paralelo, dependiendo de c칩mo est칠 configurado. Cada executor maneja su propia memoria y almacenamiento para RDDs, DataFrames, y datasets que pueden ser persistidos en memoria o en disco durante la vida de la aplicaci칩n. Los Executors son capaces de detectar fallos en las tareas y pueden reintentar la ejecuci칩n de las tareas si es necesario. Informan al Driver sobre cualquier fallo, y el Driver puede entonces decidir reprogramar la tarea en el mismo executor u otro diferente. Para mejorar el rendimiento, los executors pueden almacenar en cach칠 los datos que se utilizan frecuentemente en las tareas, evitando la necesidad de re-leer los datos del disco o recomputar los RDDs desde cero.

6. 쯈u칠 es el Catalyst Optimizer en Apache Spark y cu치l es su funci칩n principal en la optimizaci칩n de consultas SQL?abs

Es un motor de optimizaci칩n de consultas SQL extensible y altamente integrado que Spark utiliza para mejorar el rendimiento de las operaciones ejecutadas en DataFrames y Spark SQL.

La funci칩n principal del Catalyst Optimizer es transformar una expresi칩n l칩gica de consulta (por ejemplo, una secuencia de operaciones DataFrame o una consulta SQL) en un plan de ejecuci칩n f칤sico eficiente que Spark pueda ejecutar sobre el cl칰ster.

El optimizador convierte las consultas de SQL sin procesar o las operaciones de DataFrame en un 치rbol de an치lisis no resuelto. El an치lisis implica la interpretaci칩n de la consulta en t칠rminos de las estructuras de datos y operaciones internas de Spark. El 치rbol de an치lisis se transforma en un plan l칩gico resuelto, donde se resuelven las referencias a tablas, columnas y funciones. Esto incluye la aplicaci칩n de reglas que expanden las consultas, tales como el an치lisis de las subconsultas y la resoluci칩n de nombres de columnas y tablas. Luego el Catalyst aplica una serie de reglas de optimizaci칩n l칩gica para reescribir el plan l칩gico en una forma m치s eficiente. Estas reglas incluyen eliminaci칩n de subconsultas, pushdown de predicados, eliminaci칩n de proyecciones innecesarias, y otras transformaciones que reducen el coste de la consulta sin cambiar su resultado. El plan l칩gico se convierte en uno o m치s planes f칤sicos. El optimizador eval칰a diferentes estrategias de ejecuci칩n f칤sica, como diferentes formas de realizar joins (por ejemplo, broadcast hash join, sort merge join, etc.) o diferentes formas de particionamiento y distribuci칩n de datos. Finalmente, el Catalyst utiliza un modelo de coste para elegir el plan f칤sico m치s eficiente de entre las alternativas generadas. Este plan es el que finalmente se ejecuta en el cl칰ster.

Al optimizar las consultas, reduce la cantidad de datos a procesar y la complejidad de las operaciones, lo que resulta en un rendimiento significativamente mejorado y menor uso de recursos. Puede tomar decisiones basadas en el coste estimado de diferentes estrategias de ejecuci칩n, lo que ayuda a elegir el plan m치s eficiente. Est치 dise침ado para ser extensible, permitiendo a los desarrolladores a침adir sus propias reglas de optimizaci칩n y adaptar el motor a necesidades espec칤ficas.

Funci칩n de ventana en SQL para calcular la compra m치s alta realizada por cada usuario en los 칰ltimos tres meses.

```sql
SELECT DISTINCT
    `Customer ID`,
    FIRST_VALUE(`Transaction Amount`) OVER (
        PARTITION BY `Customer ID`
        ORDER BY `Transaction Amount` DESC
        RANGE BETWEEN INTERVAL 3 MONTHS PRECEDING AND CURRENT ROW
    ) AS Max_Purchase
FROM
    transactions
WHERE
    `Transaction Date` >= DATE_SUB(CURRENT_DATE(), INTERVAL 3 MONTHS)

```

````python
from pyspark.sql.functions import col, first, max, window, expr
from pyspark.sql.window import Window

# Cargar y preparar el DataFrame
df = sparksession.read.table("transactions_month")

# Filtrar los datos de los 칰ltimos tres meses
recent_df = df.filter(col("Transaction Date") >= expr("date_sub(current_date(), 90)"))

# Definir la ventana de partici칩n
windowSpec = Window.partitionBy("Customer ID").orderBy(col("Transaction Amount").desc())

# Aplicar la funci칩n de ventana
result_df = recent_df.withColumn("Max_Purchase", first("Transaction Amount").over(windowSpec))

# Seleccionar las filas 칰nicas
result_df = result_df.dropDuplicates(["Customer ID"]).select("Customer ID", "Max_Purchase")

# Mostrar el resultado
result_df.show()
````
