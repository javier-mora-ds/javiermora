1. ¬øQu√© es Apache Spark y cu√°les son sus principales ventajas sobre Pandas?

Apache Spark es un sistema de procesamiento de datos distribuido que maneja tareas de an√°lisis de datos y de ML a gran escala. Sus ventajas sobre Pandas son:

Escalabilidad. Pandas es efectivo para conjuntos de datos que caben c√≥modamente en la memoria de una sola m√°quina, Spark es dise√±ado para trabajar en un cl√∫ster de computadoras y manejar datasets de tama√±o mucho mayor que la memoria de cualquier computador individual.

Spark utiliza un motor de procesamiento en memoria, que puede ser significativamente m√°s r√°pido que las operaciones basadas en disco utilizadas por herramientas que no est√°n optimizadas para el trabajo distribuido entre varios nodos.

Facilidad para el procesamiento distribuido. Spark proporciona APIs altamente optimizadas y un planificador avanzado que facilita la escritura de aplicaciones paralelizadas eficientemente, algo que Pandas no tiene.

Soporte para m√∫ltiples lenguajes: Spark ofrece APIs en Java, Scala, Python y R, mientras que Pandas est√° limitado principalmente a Python.

Spark viene con librer√≠as integradas para diferentes tareas como SQL y DataFrames, procesamiento de gr√°ficos (GraphX), y aprendizaje autom√°tico (MLlib), lo que permite a los usuarios realizar complejas tareas de an√°lisis de datos en una √∫nica plataforma que se interconecta bien entre sus componentes.

Spark se integra con sistemas de gesti√≥n de cl√∫ster, como Hadoop YARN, Apache Mesos y Kubernetes, lo que permite una gesti√≥n eficiente de los recursos del cl√∫ster, que con Pandas no es posible.











2. ¬øQu√© es un RDD en Spark? Describe una de sus principales caracter√≠sticas. ¬øQu√© tienen que ver con los dataframes?.

Un RDD (Resilient Distributed Dataset) es la abstracci√≥n de datos primaria en Spark y proporciona una forma inmutable y distribuida de manejar los datos, permitiendo el procesamiento en m√∫ltiples nodos de un cl√∫ster. Los RDD son resistentes, lo que significa que son capaces de recomputar datos en caso de fallas.

Caracter√≠sticas de los RDD:

La resiliencia es una de sus caracter√≠sticas definitorias. Esto se logra a trav√©s de la propiedad de tolerancia a fallos, que permite que partes del dataset que se hayan perdido en el caso de un fallo en el nodo se reconstruyan autom√°ticamente. Los RDD logran esta resiliencia almacenando el linaje de las transformaciones aplicadas a los datos iniciales, permitiendo una reconstrucci√≥n eficiente de los mismos.

Inmutabilidad. Una vez que se crea un RDD, no se puede modificar. Cualquier operaci√≥n que modifique un RDD, como map o filter, resultar√° en un nuevo RDD.

Los datos en un RDD est√°n distribuidos y paralelizados autom√°ticamente a trav√©s de los nodos del cl√∫ster.

Lazy Evaluation. Spark no ejecuta las transformaciones de un RDD hasta que se realiza una acci√≥n que requiera que un resultado sea devuelto al driver.

Los RDD pueden ser almacenados en memoria durante el procesamiento, lo que permite un acceso m√°s r√°pido, especialmente √∫til para aplicaciones que acceden varias veces a los mismos datos.

Relaci√≥n entre RDD y DataFrames:

Los DataFrames en Spark son una abstracci√≥n m√°s alta que los RDD, dise√±ada para simplificar el trabajo con grandes conjuntos de datos estructurados y semiestructurados mediante una interfaz similar a la de los DataFrames de Pandas pero distribuida. Un DataFrame en Spark representa una tabla de datos con filas y columnas, donde cada columna tiene un nombre y un tipo de dato.

Los DataFrames est√°n construidos sobre el concepto de RDDs, utilizando un RDD de filas (Row RDD) donde cada fila es un objeto que representa una fila l√≥gica del DataFrame. Los DataFrames proporcionan optimizaciones a trav√©s del Catalyst query optimizer, que construye un plan de ejecuci√≥n eficiente para las consultas, algo que los RDD no hacen autom√°ticamente. Adem√°s, los DataFrames ofrecen una API muy util para la manipulaci√≥n de datos, permitiendo integraci√≥n con sistemas SQL, lo cual facilita mucho el trabajo de transformaci√≥n y an√°lisis de datos estructurados.

3. Diferencia entre transformaciones y acciones en Spark. Proporciona ejemplos de cada una. ¬øQu√© ocurre internamente cuando se ejecuta una acci√≥n?

Las transformaciones son operaciones en Spark que crean un nuevo RDD a partir de uno existente. Son lazy, es decir, no se computan inmediatamente cuando se definen, Spark construye un DAG (Directed Acyclic Graph), que es un plan de ejecuci√≥n que describe todas las transformaciones que deben ser aplicadas al RDD. Las transformaciones solo se ejecutan cuando una acci√≥n se lleva a cabo.

Ejemplos:

```python
map(func): Retorna un nuevo RDD aplicando una funci√≥n a cada elemento del RDD original.

rdd = sc.parallelize([1, 2, 3, 4])
rdd2 = rdd.map(lambda x: x * 2)

filter(func): Retorna un nuevo RDD conteniendo solo los elementos que satisfacen una condici√≥n.

rdd = sc.parallelize([1, 2, 3, 4])
filtered_rdd = rdd.filter(lambda x: x > 2)

flatMap(func): Similar a map, pero cada elemento de entrada puede ser mapeado a 0 o m√°s elementos de salida (es decir, "aplanar" la lista).

rdd = sc.parallelize(["hello world", "how are you"])
rdd2 = rdd.flatMap(lambda x: x.split(" "))
```

Las acciones son operaciones que devuelven un resultado final del procesamiento del dataset. A diferencia de las transformaciones, cuando se ejecuta una acci√≥n, Spark ejecuta el DAG creado por las transformaciones para entregar un resultado. Las acciones son necesarias para materializar los resultados del procesamiento y pueden desencadenar la computaci√≥n del RDD.

Ejemplos de Acciones:

```python
collect(): Devuelve todos los elementos del RDD como un arreglo al programa driver.

rdd = sc.parallelize([1, 2, 3, 4])
result = rdd.collect()

count(): Retorna el n√∫mero de elementos en el RDD.

rdd = sc.parallelize([1, 2, 3, 4])
count = rdd.count()

take(n): Devuelve los primeros n elementos del RDD.

rdd = sc.parallelize([1, 2, 3, 4])
first_three = rdd.take(3)

reduce(func): Agrega los elementos del RDD juntos utilizando una funci√≥n asociativa y conmutativa.

rdd = sc.parallelize([1, 2, 3, 4])
sum_result = rdd.reduce(lambda x, y: x + y)
```

Cuando se ejecuta una acci√≥n, Spark computa el RDD y sus dependencias que han sido definidos por las transformaciones:

- Eval√∫a el DAG para entender las dependencias entre las diferentes transformaciones y RDDs.
- Optimiza el plan de ejecuci√≥n, puede combinar operaciones.
- Distribuye la ejecuci√≥n del DAG a trav√©s de los nodos en el cl√∫ster para realizar el procesamiento en paralelo.
- Gestiona la recuperaci√≥n de fallos al recomputar partes perdidas del DAG si es necesario.
- Retorna los resultados al driver o escribe los resultados a un sistema de almacenamiento externo, dependiendo de la acci√≥n especificada.

El proceso de Laza evaluation y ejecuci√≥n basada en acciones permite que Spark maneje eficientemente grandes vol√∫menes de datos con un uso √≥ptimo de recursos. Esto tambi√©n ayuda a evitar repetidos c√≥mputos innecesarios y permite a Spark ejecutar optimizaciones importantes en el plan de procesamiento.

4. Explica la importancia del particionamiento en Spark y c√≥mo afecta el rendimiento del procesamiento de datos.

El particionamiento es util para el rendimiento en el procesamiento distribuido de datos. Afecta directamente c√≥mo los datos son distribuidos a trav√©s de los nodos del cl√∫ster y tiene un impacto significativo en la eficiencia de las operaciones de procesamiento paralelo.

Determina el nivel de paralelismo en las operaciones de Spark. Cada partici√≥n de un RDD, DataFrame o Dataset puede ser procesada en paralelo por diferentes nodos del cl√∫ster. Si los datos est√°n bien particionados, esto puede llevar a una utilizaci√≥n m√°s eficiente de los recursos del cl√∫ster y a una ejecuci√≥n m√°s r√°pida de las operaciones.

Un particionamiento eficaz asegura que el trabajo sea equitativamente distribuido entre todos los nodos del cl√∫ster. Esto evita que algunos nodos est√©n sobrecargados mientras que otros est√©n subutilizados, lo que puede llevar a cuellos de botella y reducir el rendimiento general del sistema.

Durante ciertas operaciones, como los joins, groupBy o reduceByKey, los datos pueden necesitar ser redistribuidos (shuffled) entre diferentes nodos para agrupar los elementos necesarios para la operaci√≥n. El shuffle es una operaci√≥n costosa porque involucra la transferencia de datos a trav√©s de la red y puede ser un factor limitante de rendimiento. Un buen particionamiento puede reducir la cantidad de datos que necesitan ser movidos, minimizando as√≠ el costo asociado con el shuffle.

EL particionamiento puede aumentar la localidad de los datos, lo que significa que los datos necesarios para procesar una tarea est√°n f√≠sicamente m√°s cerca del proceso que los necesita (idealmente en la misma m√°quina). Esto reduce la latencia y el costo de acceso a los datos, mejorando el rendimiento de las operaciones de lectura y escritura.

Para operaciones que no requieren un shuffle, como map() o filter(), un particionamiento bien dise√±ado asegura que el procesamiento se distribuya de manera uniforme, lo que maximiza el uso del paralelismo y reduce el tiempo total de ejecuci√≥n.

Para operaciones que involucran un shuffle, un particionamiento previo que alinee los datos de acuerdo con la clave sobre la cual se realizar√° la operaci√≥n (por ejemplo, usando partitionBy en un PairRDD) puede reducir significativamente la cantidad de datos que deben ser transferidos entre nodos.

Se controlar el particionamiento a trav√©s de varias configuraciones y m√©todos como repartition(), coalesce() o partitionBy()

5. ¬øCu√°les son las funciones de Spark Driver y Spark Executor?abs

El Driver es el proceso responsable de la creaci√≥n del contexto de Spark, la conversi√≥n del programa del usuario en tareas, y la planificaci√≥n de estas tareas en los executors. El Driver convierte la aplicaci√≥n Spark, que puede ser escrita en Scala, Python o Java, en un plan de ejecuci√≥n f√≠sico que se puede ejecutar a trav√©s del cl√∫ster. Esto implica la construcci√≥n del DAG (Directed Acyclic Graph) de las tareas a realizar. Programa las tareas individuales que deben ejecutarse, bas√°ndose en el DAG. Luego, env√≠a estas tareas a los executors apropiados para su ejecuci√≥n. Maneja las sesiones de Spark, incluyendo la configuraci√≥n y mantenimiento del SparkContext (en Spark 1.x) o SparkSession (en Spark 2.x y posteriores). En algunos modos de despliegue, como en standalone o Mesos, el Driver tambi√©n puede ser responsable de la gesti√≥n de recursos, negociando con el gestor de cl√∫ster para adquirir recursos para los executors. Una vez que los executors completan las tareas, env√≠an los resultados de vuelta al Driver. Este puede entonces compilar los resultados de todas las tareas para formar la salida final de la aplicaci√≥n, o realizar m√°s transformaciones y acciones seg√∫n sea necesario.

Los Executors son procesos lanzados para la duraci√≥n de la aplicaci√≥n Spark en los nodos del cl√∫ster, que ejecutan las tareas asignadas y devuelven los resultados al Driver. Sson responsables de ejecutar las tareas que les asigna el Driver. Cada executor puede ejecutar m√∫ltiples tareas en paralelo, dependiendo de c√≥mo est√© configurado. Cada executor maneja su propia memoria y almacenamiento para RDDs, DataFrames, y datasets que pueden ser persistidos en memoria o en disco durante la vida de la aplicaci√≥n. Los Executors son capaces de detectar fallos en las tareas y pueden reintentar la ejecuci√≥n de las tareas si es necesario. Informan al Driver sobre cualquier fallo, y el Driver puede entonces decidir reprogramar la tarea en el mismo executor u otro diferente. Para mejorar el rendimiento, los executors pueden almacenar en cach√© los datos que se utilizan frecuentemente en las tareas, evitando la necesidad de re-leer los datos del disco o recomputar los RDDs desde cero.

6. ¬øQu√© es el Catalyst Optimizer en Apache Spark y cu√°l es su funci√≥n principal en la optimizaci√≥n de consultas SQL?abs

Es un motor de optimizaci√≥n de consultas SQL extensible y altamente integrado que Spark utiliza para mejorar el rendimiento de las operaciones ejecutadas en DataFrames y Spark SQL.

La funci√≥n principal del Catalyst Optimizer es transformar una expresi√≥n l√≥gica de consulta (por ejemplo, una secuencia de operaciones DataFrame o una consulta SQL) en un plan de ejecuci√≥n f√≠sico eficiente que Spark pueda ejecutar sobre el cl√∫ster.

El optimizador convierte las consultas de SQL sin procesar o las operaciones de DataFrame en un √°rbol de an√°lisis no resuelto. El an√°lisis implica la interpretaci√≥n de la consulta en t√©rminos de las estructuras de datos y operaciones internas de Spark. El √°rbol de an√°lisis se transforma en un plan l√≥gico resuelto, donde se resuelven las referencias a tablas, columnas y funciones. Esto incluye la aplicaci√≥n de reglas que expanden las consultas, tales como el an√°lisis de las subconsultas y la resoluci√≥n de nombres de columnas y tablas. Luego el Catalyst aplica una serie de reglas de optimizaci√≥n l√≥gica para reescribir el plan l√≥gico en una forma m√°s eficiente. Estas reglas incluyen eliminaci√≥n de subconsultas, pushdown de predicados, eliminaci√≥n de proyecciones innecesarias, y otras transformaciones que reducen el coste de la consulta sin cambiar su resultado. El plan l√≥gico se convierte en uno o m√°s planes f√≠sicos. El optimizador eval√∫a diferentes estrategias de ejecuci√≥n f√≠sica, como diferentes formas de realizar joins (por ejemplo, broadcast hash join, sort merge join, etc.) o diferentes formas de particionamiento y distribuci√≥n de datos. Finalmente, el Catalyst utiliza un modelo de coste para elegir el plan f√≠sico m√°s eficiente de entre las alternativas generadas. Este plan es el que finalmente se ejecuta en el cl√∫ster.

Al optimizar las consultas, reduce la cantidad de datos a procesar y la complejidad de las operaciones, lo que resulta en un rendimiento significativamente mejorado y menor uso de recursos. Puede tomar decisiones basadas en el coste estimado de diferentes estrategias de ejecuci√≥n, lo que ayuda a elegir el plan m√°s eficiente. Est√° dise√±ado para ser extensible, permitiendo a los desarrolladores a√±adir sus propias reglas de optimizaci√≥n y adaptar el motor a necesidades espec√≠ficas.

Por ejemplo, UNBOUNDED PRECEDING se usa para indicar que el rango de la funci√≥n de ventana comienza desde la primera fila de la partici√≥n o del conjunto de resultados, lo cual es √∫til para calcular sumas acumulativas hasta la fila actual. - `UNBOUNDED PRECEDING` to `CURRENT ROW`: Calcula desde el inicio de la partici√≥n hasta la fila actual.
- `UNBOUNDED PRECEDING` to `UNBOUNDED FOLLOWING`: Cubre todas las filas dentro de la partici√≥n.
- `VALUE PRECEDING` to `VALUE FOLLOWING`: Establece un rango espec√≠fico basado en valores antes y despu√©s de la fila actual.

Funci√≥n de ventana en SQL para calcular la compra m√°s alta realizada por cada usuario en los √∫ltimos tres meses.

```sql
SELECT DISTINCT
    `Customer ID`,
    FIRST_VALUE(`Transaction Amount`) OVER (
        PARTITION BY `Customer ID`
        ORDER BY `Transaction Amount` DESC
        RANGE BETWEEN INTERVAL 3 MONTHS PRECEDING AND CURRENT ROW
    ) AS Max_Purchase
FROM
    transactions
WHERE
    `Transaction Date` >= DATE_SUB(CURRENT_DATE(), INTERVAL 3 MONTHS)

```

````python
# C√≥digo Aqu√≠

import pickle

# Cargar el objeto serializado desde un archivo
with open('object.pkl', 'rb') as file:
    model = pickle.load(file)

# Mostrar informaci√≥n sobre el objeto
print(dir(model))
````

````python
# Mostrar la documentaci√≥n del objeto
help(model)
````

````python
from pyspark.sql.functions import udf
from pyspark.sql.types import FloatType

# usando el m√©todo predict_proba que devuelve la probabilidad de la clase positiva
def predict_probability(amount, quantity, age, hour):
    # Verificar si alguno de los inputs es None
    if amount is None or quantity is None or age is None or hour is None:
        return None
    # Asumimos que el modelo espera una lista con los valores en este orden espec√≠fico
    features = [[amount, quantity, age, hour]]
    # predict_proba devuelve un array de [probabilidad_clase_negativa, probabilidad_clase_positiva]
    try:
        probability = float(model.predict_proba(features)[0][1])
        return probability
    except Exception as e:
        # Log or handle the exception in an appropriate way
        return None

# Registrar la UDF
udf_predict_probability = udf(predict_probability, FloatType())
````

````python

df_with_predictions = df.withColumn("fraud_probability", udf_predict_probability(col("Transaction Amount"), col("Quantity"), col("Customer Age"), col("Transaction Hour")))

# Mostrar algunos resultados
df_with_predictions.show()
````

````python
def fraud_probability(amount, quantity, age, hour):
    # Asumimos que el modelo espera una lista con estos valores
    features = [[amount, quantity, age, hour]]
    # El modelo predice las probabilidades y seleccionamos la √∫ltima de la lista, que asumimos es la de fraude
    try:
        # Obtenemos la √∫ltima probabilidad, que corresponde a la probabilidad de fraude
        fraud_prob = model.predict_proba(features)[-1][-1]
        return float(fraud_prob)
    except Exception as e:
        # Puedes decidir qu√© hacer en caso de excepci√≥n, aqu√≠ simplemente retornamos None
        print(f"Error al predecir: {e}")
        return None

# Registrar la UDF en Spark
udf_fraud_probability = udf(fraud_probability, FloatType())
````

````python
df_with_fraud_prob = df.withColumn("Fraud Probability", udf_fraud_probability(
    col("Transaction Amount"),
    col("Quantity"),
    col("Customer Age"),
    col("Transaction Hour")
))

# Mostrar los resultados para verificar
df_with_fraud_prob.show()
````

````python
# A√±adir una nueva columna 'prediction' utilizando la UDF
df_with_prediction = df.withColumn("prediction", udf_fraud_probability(
    col("Transaction Amount"),
    col("Quantity"),
    col("Customer Age"),
    col("Transaction Hour")
))

# Mostrar los resultados para verificar
df_with_prediction.select("Customer ID", "Transaction Amount", "prediction").show()
````
